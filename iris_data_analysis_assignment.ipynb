### [Problem 4] Investigation of the iris dataset itself

---





*   It is a multivariate data set introduced by the British statistician, eugenicist, and biologist Ronald Fisher in his 1936 paper The use of multiple measurements in taxonomic problems as an example of linear discriminant analysis.
*   The data set consists of 50 samples from each of three species of Iris (Iris setosa, Iris virginica and Iris versicolor). Four features were measured from each sample: the length and the width of the sepals and petals, in centimeters.
*  Based on Fisher's linear discriminant model, this data set became a typical test case for many statistical classification techniques in machine learning such as support vector machines.

### [Question 6] Creating a diagram

---





*   `Make a pie chart of the number of samples per label (displayed as a percentage):`



```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sr = df["Species"].value_counts()
labels = 'setosa', 'virgiclor','virginca'
plt.title("A pie chart of the number of samples per label")
sr.plot.pie(autopct="%.1f%%", labels=labels);
```



*   `Select features one by one and use a boxplot to visualize the distribution of data for each label:`



```python
fig, axes = plt.subplots(1, 2)

sns.boxplot(x="Species", y="petal_length",data=df, ax=axes[0])
sns.boxplot(x="Species", y="sepal_length",data=df, ax=axes[1])

axes[0].set_xticklabels(['setosa','virgiclor','virginica'])
axes[0].set_title('petal_length boxplot')

axes[1].set_xticklabels(['setosa','virgiclor','virginica'])
axes[1].set_title('sepal_length boxplot')

fig.set_size_inches(14, 5)
plt.show()
```



*   `Select features one by one and visualize the distribution of data for each label using a violin diagram:`



```python
fig, axes = plt.subplots(1, 2)

sns.violinplot(x="Species", y="sepal_length",data=df,ax=axes[0])
sns.violinplot(x="Species", y="petal_length",data=df,ax=axes[1])

axes[0].set_xticklabels(['setosa','virgiclor','virginica'])
axes[0].set_title('sepal_length violinplot')

axes[1].set_xticklabels(['setosa','virgiclor','virginica'])
axes[1].set_title('petal_length violinplot')

fig.set_size_inches(14, 5)
plt.show()
```

A **violin plot** is more informative than a plain **box plot**. While a **box plot** only shows summary statistics such as mean/median and interquartile ranges, the **violin plot** shows the full distribution of the data.

### [Problem 7] Checking the relationship between feature values

---



* `Create scatter plots of all combinations (6) such that 2 out of 4 features are selected by color:`

```python
fig, ax = plt.subplots(2, 3, figsize=(15,10))

fig.suptitle('- Create scatter plots of all combinations (6) such that 2 out of 4 features are selected by color.')

sns.scatterplot(ax=ax[0,0], data=df, x='sepal_length', y='sepal_width', hue='Species', palette="deep")
sns.scatterplot(ax=ax[0,1], data=df, x='sepal_length', y='petal_length', hue='Species', palette="deep")
sns.scatterplot(ax=ax[0,2], data=df, x='sepal_length', y='petal_width', hue='Species', palette="deep")
sns.scatterplot(ax=ax[1,0], data=df, x='sepal_width', y='petal_length', hue='Species', palette="deep")
sns.scatterplot(ax=ax[1,1], data=df, x='sepal_width', y='petal_width', hue='Species', palette="deep")
sns.scatterplot(ax=ax[1,2], data=df, x='petal_length', y='petal_width', hue='Species', palette="deep")

sns.set()
```

* `Create a scatterplot matrix that displays all feature combinations at once:`

```python
rel=sns.pairplot(df, hue='Species')

rel.fig.subplots_adjust(top=.9)
rel.fig.suptitle('- Create a scatterplot matrix that displays all feature combinations at once.')

sns.set()
```
